```python
DefineCohortObject = iceesclient.DefineCohort()
```

```python
define_a_cohort = DefineCohortObject.run_define_cohort('AvgDailyPM2.5Exposure', '3','>')
```

```python
define_a_cohort['return value']
```

```python
CohortDefinitionObject = iceesclient.GetCohortDefinition()
```

```python
cohort_definition = CohortDefinitionObject.run_get_cohort_definition('COHORT:45')
```

```python
cohort_definition['return value']
```
